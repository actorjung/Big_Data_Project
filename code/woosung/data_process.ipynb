# 2. 광진구 공영주차장 정보

# 2-1. 전처리

```python
#nan값 행 확인
df1[df1["주차요금 종일"].isnull()]
```

```python
#nan값 0으로 바꾸기
df1["부지면적"] = df1["부지면적"].fillna(0)
df1["주차면수 노상"] = df1["주차면수 노상"].fillna(0)
df1["주차면수 노외"] = df1["주차면수 노외"].fillna(0)
df1["주차요금 5분당"] = df1["주차요금 5분당"].fillna(0)
df1["주차요금 주간"] = df1["주차요금 주간"].fillna(0)
df1["주차요금 야간"] = df1["주차요금 야간"].fillna(0)
df1["주차요금 종일"] = df1["주차요금 종일"].fillna(30000)
```

```python
#nan값 다른 문자로 바꾸기
df1["거주자"] = df1["거주자"].fillna('X')
df1["공영"] = df1["공영"].fillna('X')
df1["시설규모"] = df1["시설규모"].fillna('1층1단')
```

```python
#(만원)삭제하기, (원)삭제하기
df1['주차요금 5분당'] = df1['주차요금 5분당'].str.replace('원', '')
df1['주차요금 주간'] = df1['주차요금 주간'].str.replace('만원', '0000')
df1['주차요금 야간'] = df1['주차요금 야간'].str.replace('만원', '0000')
df1['주차요금 종일'] = df1['주차요금 종일'].str.replace('만원', '0000')
```

```python
df1.to_csv(f"/{Path2}/서울특별시 광진구_공영주차장 정보_20220118.csv", index = False)
```
