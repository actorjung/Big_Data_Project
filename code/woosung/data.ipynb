```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import normaltest
import scipy.stats as stats
import seaborn as sns
```

```python
# 정규성 검정
for i in df.columns[1:]:
    statistic, p_value = normaltest(df[i])
    if p_value < 0.05:
        print(f"{i} 데이터는 정규 분포가 아닙니다.")
    else:
        print(f"{i} 데이터는 정규 분포입니다.")
```

```python
for i in df.columns[1:]:
    plt.hist(df[i], bins=20, density=True, alpha=0.6, color='g')

    # Fit a normal distribution to the data
    mu, std = stats.norm.fit(df[i])
    xmin, xmax = plt.xlim()
    x = np.linspace(xmin, xmax, 100)
    p = stats.norm.pdf(x, mu, std)
    plt.plot(x, p, 'k', linewidth=2)

    # Add a title and axis labels
    plt.title("Histogram and Normal Probability Plot")
    plt.xlabel("Value")
    plt.ylabel("Density")

    # Create a normal probability plot
    plt.figure()
    stats.probplot(df[i], plot=plt)

    # Display the plot
    plt.show()

    # 히스토그램: 히스토그램은 무작위 데이터의 분포를 보여줍니다. 각 막대는 값의 범위(구간)를 나타내고 막대의 높이는 해당 구간에 속하는 데이터 포인트의 수를 나타냅니다. 히스토그램은 정규 분포가 종 모양의 곡선을 갖는 경향이 있으므로 데이터가 정규 분포를 따르는지 여부를 시각적으로 표시할 수 있습니다.
    # 정규 확률도: 정규 확률도는 데이터가 정규 분포를 따르는지 여부를 그래픽으로 평가하는 방법입니다. 평균 및 표준 편차가 동일한 정규 분포의 분위수에 대해 데이터의 분위수를 플로팅합니다. 데이터가 정규 분포를 따르는 경우 플롯의 점은 대략 직선을 형성해야 합니다. 점이 직선에서 크게 벗어나면 데이터가 정규 분포를 따르지 않는다는 것을 나타냅니다.
```
