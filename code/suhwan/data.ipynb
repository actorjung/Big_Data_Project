```python
import pandas as pd
import os
```

```python
folder_path = 'D:/Big_Data_Project/data/'
```

```python
df1=pd.read_csv(f'{folder_path}2017_2019_도로교통공단_광진구_행정동별_교통사고_현황.csv')
df2=pd.read_csv(f'{folder_path}LED바닥형 신호등.csv')
df3=pd.read_csv(f'{folder_path}공영주차장.csv')
df4=pd.read_csv(f'{folder_path}무인교통단속카메라.csv')
df5=pd.read_csv(f'{folder_path}서울시 광진구 안심이 CCTV 연계 현황.csv')
#df6=pd.read_csv(f'{folder_path}서울시 광진구 초등학교 기본정보.csv')
df7=pd.read_csv(f'{folder_path}서울시 범죄지도를 이용한 광진구 범죄율.csv')
df8=pd.read_csv(f'{folder_path}서울특별시 광진구_불법주정차 위반 단속 CCTV 위치정보_20220502.csv')
df9=pd.read_csv(f'{folder_path}시설관리공단 공영주차장.csv')
df10=pd.read_csv(f'{folder_path}어린이보호구역.csv')
df11=pd.read_csv(f'{folder_path}어린이집.csv')
df12=pd.read_csv(f'{folder_path}인구현황.csv')
# + 면적 관련 데이터
```

```python
df1.rename(columns={'법정동명':'행정동'},inplace=True)
```

```python
df1
```

```python
df2.rename(columns={'동명':'행정동'},inplace=True)
```

```python
df2=df2.groupby('행정동').agg(신호등수=('수량','sum')).reset_index()
```

```python
df2
```

```python
df3['주차요금 5분당'].unique()
```

```python
df3
```

```python
df3['주차요금 5분당']=df3['주차요금 5분당'].replace('250원','250').replace('200원','200').replace('75원','75').replace('150원','150').replace('75원 ','75')
```

```python
df3['주차요금 5분당']=df3['주차요금 5분당'].astype(int)
```

```python
df3.rename(columns={'동명':'행정동'},inplace=True)
```

```python
df3=df3.groupby('행정동').agg(주차장수=('행정동','count'),주차면수소계=('주차면수 소계','sum'),주차요금5분=('주차요금 5분당','sum')).reset_index()
df3
```

```python
df4.rename(columns={'동명':'행정동'},inplace=True)
```

```python
df4=df4.groupby('행정동').agg(무인교통단속카메라수량=('행정동','count')).reset_index()
```

```python
df4
```

```python
df5=df5.groupby('행정동').agg(CCTV수량=('행정동','count')).reset_index()
df5
```

```python
df7.rename(columns={'ADM_DR_NM':'행정동'},inplace=True)
df7.rename(columns={'평균':'범죄율평균'},inplace=True)
```

```python
df7
```

```python
df8=df8.groupby('행정동').agg(불법주정차단속CCTV수량=('행정동','count')).reset_index()
df8
# 행정동 나누기
```

```python
df9.rename(columns={'동명':'행정동'},inplace=True)
```

```python
df9.columns
```

```python
df9['행정동'].unique()
```

```python
df9=df9.groupby('행정동').agg(시설관리공단주차장수=('행정동','count'),주차면수=('주차면수','sum'),주차요금1시간=(' 시간당주차요금 ','sum')).reset_index()
df9
```

```python
df10.rename(columns={'동명':'행정동'},inplace=True)
```

```python
df10=df10.groupby('행정동').agg(어린이보호구역=('행정동','count')).reset_index()
df10
```

```python
df11.rename(columns={'동명':'행정동'},inplace=True)
```

```python
df11.columns
```

```python
df11=df11.groupby('행정동').agg({'정원수':'sum'}).reset_index()
df11
```

```python
df12=df12.drop(0).reset_index(drop=True)
```

```python
df12.rename(columns={'행정기관':'행정동'},inplace=True)
```

```python
df = pd.merge(df1,df2,how='outer', on='행정동')
df = pd.merge(df,df3,how='outer', on='행정동')
df = pd.merge(df,df4,how='outer', on='행정동')
df = pd.merge(df,df5,how='outer', on='행정동')
df = pd.merge(df,df7,how='outer', on='행정동')
df = pd.merge(df,df9,how='outer', on='행정동')
df = pd.merge(df,df10,how='outer', on='행정동')
df = pd.merge(df,df11,how='outer', on='행정동')
df = pd.merge(df,df12,how='outer', on='행정동')
```
