```python
#nan값 0으로 바꾸기
df1["부지면적"] = df1["부지면적"].fillna(0)
df1["주차면수 노상"] = df1["주차면수 노상"].fillna(0)
df1["주차면수 노외"] = df1["주차면수 노외"].fillna(0)
df1["주차요금 5분당"] = df1["주차요금 5분당"].fillna(0)
df1["주차요금 주간"] = df1["주차요금 주간"].fillna(0)
df1["주차요금 야간"] = df1["주차요금 야간"].fillna(0)
df1["주차요금 종일"] = df1["주차요금 종일"].fillna(30000)
```
