# 2. 광진구 공영주차장 정보

```python
#맥북은 인코딩 방식을 utf-8사용하기 때문에 윈도우 방식인 cp949로 인코딩을 해준다.
df1 = pd.read_csv(f"/{Path}/서울특별시 광진구_공영주차장 정보_20220118.csv",encoding = 'cp949')
```

```python
df1
```

# 2-1. 전처리

```python
#nan값 행 확인
df1[df1["주차요금 종일"].isnull()]
```

```python
#nan값 0으로 바꾸기
df1["부지면적"] = df1["부지면적"].fillna(0)
df1["주차면수 노상"] = df1["주차면수 노상"].fillna(0)
df1["주차면수 노외"] = df1["주차면수 노외"].fillna(0)
df1["주차요금 5분당"] = df1["주차요금 5분당"].fillna(0)
df1["주차요금 주간"] = df1["주차요금 주간"].fillna(0)
df1["주차요금 야간"] = df1["주차요금 야간"].fillna(0)
df1["주차요금 종일"] = df1["주차요금 종일"].fillna(0)
```

```python
#nan값 다른 문자로 바꾸기
df1["거주자"] = df1["거주자"].fillna('X')
df1["공영"] = df1["공영"].fillna('X')
df1["시설규모"] = df1["시설규모"].fillna('1층1단')
```

```python
#(만원)삭제하기, (원)삭제하기
df1['주차요금 5분당'] = df1['주차요금 5분당'].str.replace('원', '')
df1['주차요금 주간'] = df1['주차요금 주간'].str.replace('만원', '0000')
df1['주차요금 야간'] = df1['주차요금 야간'].str.replace('만원', '0000')
df1['주차요금 종일'] = df1['주차요금 종일'].str.replace('만원', '0000')
```

```python
df1
```

```python
df1.to_csv(f"/{Path2}/서울특별시 광진구_공영주차장 정보_20220118.csv", index = False)
```

# 2-2 . 필요없는 (속성)열 제거

# 4. 광진구 야간개방 주차장 

```python
df3 = pd.read_csv(f"/{Path}/서울특별시 광진구_야간개방주차장_20221227.csv", encoding = 'utf-8')
```

```python
df3
```

# 4-1. 전처리 

```python
#ho-호,road_nm-도로명이름,bld_num-건물번호,resident_day_month_amt-거주자 주간 월금액,
#resident_night_month_amt-거주자 야간 월금액,resident_allday_month_amt-거주자-전일 월금액,specific-특이사항
```

```python
#nan값 행 확인
df3[df3["specific"].isnull()]
```

```python
#nan값 0으로 바꾸기
df3["resident_day_month_amt"] = df3["resident_day_month_amt"].fillna(0)
df3["resident_night_month_amt"] = df3["resident_night_month_amt"].fillna(0)
df3["resident_allday_month_amt"] = df3["resident_allday_month_amt"].fillna(0)
```

```python
#nan값 다른 문자로 바꾸기
df3["ho"] = df3["ho"].fillna('X')
df3["road_nm"] = df3["road_nm"].fillna('X')
df3["bld_num"] = df3["bld_num"].fillna('X')
df3["specific"] = df3["specific"].fillna('X')
```

```python
df3
```

```python
df3.to_csv(f"/{Path2}/서울특별시 광진구_야간개방주차장_20221227.csv", index = False)
```
