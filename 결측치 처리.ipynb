```python
import pandas as pd
import os
```

# 전통시장

```python
dirpath='data/서울특별시 광진구_전통시장 주변 주차 최신화_20221125/'
files=os.listdir(dirpath)
files
```

```python
raw = pd.DataFrame()

# 반복문 -> csv 파일 읽기 -> 데이터프레임에 추가하기
for file in files:
    #print(dirpath + file)
    temp = pd.read_excel(dirpath + file)
    raw = pd.concat([raw, temp], axis=0,ignore_index=True)
```

```python
raw
```

```python
raw=raw[['전통시장명(market_nm)','시장 종류(market_type)','시장과의 거리(m)(distance)','주차장고유값(pl_id)',
         '주차장이름(pl_nm)','법정동코드(bemd_code)','유무료(pl_fee)','일반-기본단위시간(pl_basic_rate_min)','일반-기본단위금액(pl_basic_rate_amt)',
         '일반-추가단위시간(pl_addi_rate_min)','일반-추가단위금액(pl_addi_rate_amt)','일반-1일(전일)(pl_allday_pass_amt)','계획주차면수(pl_num_space_planned)',
         '경도(logitude)','위도(latitude)']]
```

```python
raw.isnull().sum()
```

```python
raw.dropna(inplace=True)
```

```python
raw=raw[raw['계획주차면수(pl_num_space_planned)']!=0]
```
