```python
dirpath='data/서울특별시 광진구_전통시장 주변 주차 최신화_20221125/'
files=os.listdir(dirpath)
files
```

```python
raw = pd.DataFrame()

# 반복문 -> csv 파일 읽기 -> 데이터프레임에 추가하기
for file in files:
    print(dirpath + file)
    temp = pd.read_excel(dirpath + file)
    raw = pd.concat([raw, temp], axis=0,ignore_index=True)
```
